```python
reservas = {}

def ReservarQuarto():
    print("Qual quarto deseja?")
    print("Basico - R$200")
    print("Medio - R$500")
    print("Vip - R$1000")
    nome = input("Digite seu nome (Seu nome é seu registro de enrada)")
    quarto = input("Escreva qual nível de quarto quer (Ex.: 'Basico')").lower()
    dias = int(input("Quantos dias de estadia?"))
    if quarto != "basico" or quarto != "medio" or quarto != "vip":
        print("Nao temos esse quarto")
    else:
        reservas[nome] = [quarto, dias]
        print("Quarto reservado.")

def RemoverReserva():
    remover = input("Digite seu registo de entrada para remover sua reserva.")
    reservas.pop(remover)
    print("Reserva removida")

def ListarReservas():
    for nome, quarto in reservas.items():
        print(f"O hospede {nome} reservou um quarto {quarto[0]} por {quarto[1]} dias!!")

def Custo():
    print("Para saber o Custo da sua hospedagem digite seu registo")
    registro = input("Qual seu nome de registro")
    
   
    if registro in reservas:
        if reservas[registro][0] == "basico":
            print("Sua reserva é de R$200 reais")
        elif reservas[registro][0] == "medio":
            print("Sua reserva é de R$500 reais")
        else:
            print("Sua reserva é de R$1000 reais")
    else:
        print("Registro não encontrado.")

ReservarQuarto()
ListarReservas()
RemoverReserva()
Custo()
```

```python
estacionamento = {}
def AddCarro():
    placa = input("Digite sua placa: ")
    marca = input("Qual marca do seu carro: ")
    cor = input("Qual a cor do seu carro: ")
    
    estacionamento[placa] = marca, cor
    
def RemoverCarro():
    sair = input("Digite a placa do carro que esta saindo")
    estacionamento.pop(sair)
    
def ListarCarros():
    for placa, lista in estacionamento.items():
        print(f"Marca: {lista[0]}\nCor: {lista[1]}\nPlaca: {placa}")
AddCarro()
ListarCarros()
```

```python
projetos = {}

def AdicionarProjeto(nome_projeto):
    if nome_projeto not in projetos:
        projetos[nome_projeto] = []
        print(f"Projeto '{nome_projeto}' adicionado com sucesso.")
    else:
        print(f"Projeto '{nome_projeto}' já existe.")

def AdicionarTarefa(nome_projeto, titulo, descricao, responsavel, status):
    if nome_projeto in projetos:
        tarefa = {
            "titulo": titulo,
            "descricao": descricao,
            "responsavel": responsavel,
            "status": status
        }
        projetos[nome_projeto].append(tarefa)
        print(f"Tarefa '{titulo}' adicionada ao projeto '{nome_projeto}'.")
    else:
        print(f"Projeto '{nome_projeto}' não encontrado.")

def AtualizarTarefa(nome_projeto, titulo_tarefa, novo_status):
    if nome_projeto in projetos:
        for tarefa in projetos[nome_projeto]:
            if tarefa["titulo"] == titulo_tarefa:
                tarefa["status"] = novo_status
                print(f"Tarefa '{titulo_tarefa}' atualizada para '{novo_status}'.")
                return
        print(f"Tarefa '{titulo_tarefa}' não encontrada no projeto '{nome_projeto}'.")
    else:
        print(f"Projeto '{nome_projeto}' não encontrado.")

def ListarTarefas(nome_projeto):
    if nome_projeto in projetos:
        print(f"Tarefas do projeto '{nome_projeto}':")
        for tarefa in projetos[nome_projeto]:
            print(f"- Título: {tarefa['titulo']}, Descrição: {tarefa['descricao']}, "
                  f"Responsável: {tarefa['responsavel']}, Status: {tarefa['status']}")
    else:
        print(f"Projeto '{nome_projeto}' não encontrado.")
AdicionarProjeto("Projeto A")
AdicionarTarefa("Projeto A", "Tarefa 1", "Descrição da tarefa 1", "Carlos", "em andamento")
AdicionarTarefa("Projeto A", "Tarefa 2", "Descrição da tarefa 2", "Henrique", "nao iniciei")
ListarTarefas("Projeto A")
AtualizarTarefa("Projeto A", "Tarefa 1", "concluído")
ListarTarefas("Projeto A")
```
